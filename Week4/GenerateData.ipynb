```python

X, y = make_blobs(n_samples=1000, centers=10, n_features=64)
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
figure = plt.figure(figsize=(7, 3.5))
rows, cols= 2, 5
for k in range(10):
    img = X[k, :].reshape(5, 5)
    figure.add_subplot(rows, cols, k+1)
    plt.axis("off")
    plt.imshow(img, cmap="gray")
plt.show()
```
