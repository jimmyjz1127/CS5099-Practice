We want to compute the intractable GMM posterior:
$$
    p(\mathbf{z},\boldsymbol{\mu}|\mathbf{X}) = \frac{p(\mathbf{X},\boldsymbol{\mu},\mathbf{z})}{p(\mathbf{X})}
$$

Introduce approximating distribution $q(\mathbf{z},\boldsymbol{\mu})$ where we use KL divergence as a cost function 
$$
    \mathbb{KL}(q||p) = \sum_{\mathbf{z},\boldsymbol{\mu}}q(\mathbf{z},\boldsymbol{\mu})\log\frac{q(\mathbf{z},\boldsymbol{\mu})}{p(\mathbf{z},\boldsymbol{\mu}|\mathbf{X})}
$$

Re-write as objective function
$$\begin{align}
    \mathcal{J}(q) &= \mathbb{KL}(q||p) = \int\sum_{\mathbf{z}}q(\mathbf{z},\boldsymbol{\mu})\log\frac{q(\mathbf{z},\boldsymbol{\mu})}{p(\mathbf{z},\boldsymbol{\mu}|\mathbf{X})}d\boldsymbol{\mu}\\

        &=  \int\sum_{\mathbf{z}}q(\mathbf{z},\boldsymbol{\mu})\log q(\mathbf{z},\boldsymbol{\mu})d\boldsymbol{\mu} - \int\sum_{\mathbf{z}}q(\mathbf{z},\boldsymbol{\mu})\log p(\mathbf{z},\boldsymbol{\mu}|\mathbf{X})d\boldsymbol{\mu}\\

        &= \int\sum_{\mathbf{z}}q(\mathbf{z},\boldsymbol{\mu})\log q(\mathbf{z},\boldsymbol{\mu})d\boldsymbol{\mu} -\int\sum_{\mathbf{z}}q(\mathbf{z},\boldsymbol{\mu})\log \frac{p(\mathbf{z},\boldsymbol{\mu},\mathbf{X})}{p(\mathbf{X})}d\boldsymbol{\mu}\\

        &= \int\sum_{\mathbf{z}}q(\mathbf{z},\boldsymbol{\mu})\log q(\mathbf{z},\boldsymbol{\mu})d\boldsymbol{\mu} - \int\sum_{\mathbf{z}}q(\mathbf{z},\boldsymbol{\mu})\log p(\mathbf{z},\boldsymbol{\mu},\mathbf{X})d\boldsymbol{\mu} + \int\sum_{\mathbf{z}}q(\mathbf{z},\boldsymbol{\mu})\log p(\mathbf{X})d\boldsymbol{\mu}\\

        &= \int\sum_{\mathbf{z}}q(\mathbf{z},\boldsymbol{\mu})\log q(\mathbf{z},\boldsymbol{\mu})d\boldsymbol{\mu} - \int\sum_{\mathbf{z}}q(\mathbf{z},\boldsymbol{\mu})\log p(\mathbf{z},\boldsymbol{\mu},\mathbf{X})d\boldsymbol{\mu} + p(\mathbf{X})\\

        &= \int\sum_{\mathbf{z}}q(\mathbf{z},\boldsymbol{\mu})\log \frac{q(\mathbf{z},\boldsymbol{\mu})}{p(\mathbf{z},\boldsymbol{\mu},\mathbf{X})}d\boldsymbol{\mu} + p(\mathbf{X})\\

        &= \mathbb{KL}\Big(q(\mathbf{z},\boldsymbol{\mu}) || p(\mathbf{z},\boldsymbol{\mu},\mathbf{X})\Big) + p(\mathbf{X})
\end{align}$$

Note 
$$
    \mathcal{J}(q) = \mathbb{KL}\Big(q(\mathbf{z},\boldsymbol{\mu}) || p(\mathbf{z},\boldsymbol{\mu},\mathbf{X})\Big) + \underbrace{p(\mathbf{X})}_{\text{constant}}
$$

Hence our objective can simplify to
$$\begin{align}
    q* &\leftarrow \argmin_{q} \left[ \mathbb{KL}\Big(q(\mathbf{z},\boldsymbol{\mu}) || p(\mathbf{z},\boldsymbol{\mu},\mathbf{X})\Big) + p(\mathbf{X}) \right]\\
        &= \argmin_{q} \left[ \mathbb{KL}\Big(q(\mathbf{z},\boldsymbol{\mu}) || p(\mathbf{z},\boldsymbol{\mu},\mathbf{X})\Big) \right]
\end{align}$$

Define new objective function as $\mathcal{L}(q) = -\mathcal{J}(q)$

We can show that $\mathcal{L}(q)$ is ELBO for data log likelihood :
$$\begin{align}
    \log p(\mathbf{X}) = \log \int \sum_z p(\mathbf{X},\mathbf{z},\boldsymbol{\mu})d\mu
\end{align}$$
The following always holds
$$\begin{align}
    \log p(\mathbf{X}) = \log \int \sum_z q(\boldsymbol{\mu},\mathbf{z}) \cdot \frac{p(\mathbf{X},\mathbf{z},\boldsymbol{\mu})}{q(\boldsymbol{\mu},\mathbf{z})} d\mu   
\end{align}$$    
Jensen's Inequality gives the following
$$\begin{align}
    &\log p(\mathbf{X}) = \log \int \sum_z q(\boldsymbol{\mu},\mathbf{z}) \cdot \frac{p(\mathbf{X},\mathbf{z},\boldsymbol{\mu})}{q(\boldsymbol{\mu},\mathbf{z})} d\mu \geq \int \sum_z q(\boldsymbol{\mu},\mathbf{z}) \cdot \log \frac{p(\mathbf{X},\mathbf{z},\boldsymbol{\mu})}{q(\boldsymbol{\mu},\mathbf{z})} d\mu\\\\

    &\implies \log p(\mathbf{X}) \geq \int \sum_z q(\boldsymbol{\mu},\mathbf{z}) \cdot \log \frac{p(\mathbf{X},\mathbf{z},\boldsymbol{\mu})}{q(\boldsymbol{\mu},\mathbf{z})} d\mu\\\\

    &\implies \log p(\mathbf{X}) \geq -\mathbb{KL}\Big(q(\mathbf{z},\boldsymbol{\mu}) || p(\mathbf{z},\boldsymbol{\mu},\mathbf{X})\Big)\\\\

    &\implies \log p(\mathbf{X}) \geq \mathcal{L}(q)
\end{align}$$



Goal is to maximize log marginal likelihood : 
$$
    \log p(\mathbf{X}) = \log \int \sum_z p(\mathbf{X},\mathbf{z},\boldsymbol{\mu})d\mu
$$
which is intractible

Introduce Approximating Distribution : $q(\boldsymbol{\mu},\mathbf{z})$  
$$\begin{align}
    \log p(\mathbf{X}) = \log \int \sum_z q(\boldsymbol{\mu},\mathbf{z}) \cdot \frac{p(\mathbf{X},\mathbf{z},\boldsymbol{\mu})}{q(\boldsymbol{\mu},\mathbf{z})} d\mu
\end{align}$$

Jensen's Inequality gives us
$$\begin{align}
    &\log p(\mathbf{X}) = \log \int \sum_z q(\boldsymbol{\mu},\mathbf{z}) \cdot \frac{p(\mathbf{X},\mathbf{z},\boldsymbol{\mu})}{q(\boldsymbol{\mu},\mathbf{z})} d\mu \geq \int \sum_z q(\boldsymbol{\mu},\mathbf{z}) \cdot \log \frac{p(\mathbf{X},\mathbf{z},\boldsymbol{\mu})}{q(\boldsymbol{\mu},\mathbf{z})} d\mu\\\\

    &\implies \log p(\mathbf{X}) \geq \int \sum_z q(\boldsymbol{\mu},\mathbf{z}) \cdot \log \frac{p(\mathbf{X},\mathbf{z},\boldsymbol{\mu})}{q(\boldsymbol{\mu},\mathbf{z})} d\mu\\\\

    &\implies \log p(\mathbf{X}) \geq -\mathbb{KL}\Big(q(\boldsymbol{\mu},\mathbf{z})||p(\mathbf{X},\mathbf{z},\boldsymbol{\mu})\Big)\\\\

    &\implies \log p(\mathbf{X}) \geq \mathbb{E}_{q(\mu,z)}\left[ \frac{p(\mathbf{X},\mathbf{z},\boldsymbol{\mu})}{q(\boldsymbol{\mu},\mathbf{z})} \right] = \mathbb{E}_{q(\mu,z)}[\log p(\mathbf{X},\mathbf{z},\boldsymbol{\mu})] - \mathbb{E}_{q(\mu,z)}[\log q(\mu,z)]
\end{align}$$
So,
$$
    \boxed{\mathcal{L}(q) = \mathbb{E}_{q(\mu,z)}[\log p(\mathbf{X},\mathbf{z},\boldsymbol{\mu})] - \mathbb{E}_{q(\mu,z)}[\log q(\mu,z)]}
$$

Target posterior : 
$$
    p(\mathbf{z},\boldsymbol{\mu} | \mathbf{X}) = \frac{p(\mathbf{X},\mathbf{z},\boldsymbol{\mu})}{p(\mathbf{X})}\\
    \implies p(\mathbf{X},\mathbf{z},\boldsymbol{\mu}) = p(\mathbf{z},\boldsymbol{\mu} | \mathbf{X}) \cdot p(\mathbf{X})
$$
Hence,
$$ \begin{align}
    \mathcal{L}(q) &= \mathbb{E}_{q(\mu,z)}[\log p(\mathbf{X},\mathbf{z},\boldsymbol{\mu})] - \mathbb{E}_{q(\mu,z)}[\log q(\mu,z)]\\\\
                   &= \mathbb{E}_{q(\mu,z)}[\log p(\mathbf{z},\boldsymbol{\mu} | \mathbf{X}) p(\mathbf{X})] - \mathbb{E}_{q(\mu,z)}[\log q(\mu,z)]\\\\
                   &= \mathbb{E}_{q(\mu,z)}[\log p(\mathbf{z},\boldsymbol{\mu} | \mathbf{X})] + \log p(\mathbf{X})- \mathbb{E}_{q(\mu,z)}[\log q(\mu,z)]\\\\
                   &= \mathbb{E}_{q(\mu,z)}\left[\log \frac{p(\mathbf{z},\boldsymbol{\mu} | \mathbf{X})}{q(\mu,z)}\right] + \log p(\mathbf{X})\\\\
                   &= -\mathbb{KL}(q(\mathbf{z},\boldsymbol{\mu}) || p(\mathbf{z},\boldsymbol{\mu} | \mathbf{X})) + \underbrace{\log p(\mathbf{X})}_{\text{constant}}\\\\

\end{align}$$
This shows that to maximize the lower bound, we want our proposed distribution to approximate the posterior $p(\mathbf{z},\boldsymbol{\mu} | \mathbf{X})$
